```python
import yfinance as yf
import pandas as pd

# Fetch historical stock data for Starbucks
starbucks = yf.Ticker("SBUX")
starbucks_data = starbucks.history(period="5y")

# Fetch financials (income statement, balance sheet, and cash flow)
income_statement = starbucks.financials
balance_sheet = starbucks.balance_sheet
cash_flow = starbucks.cashflow

# Display the first few rows of each dataset
print("Income Statement:")
print(income_statement.head())

print("\nBalance Sheet:")
print(balance_sheet.head())

print("\nCash Flow Statement:")
print(cash_flow.head())
```

IMPORTANT NOTE

"yf.Ticker("SBUX")" initializes the Ticker object for Starbucks.

".history(period="5y")" retrieves historical stock data for the last five years.

".financials, .balance_sheet, and .cashflow" fetch the respective financial statements.

1. Revenue and Profit Trends:

```python
# Extract revenue and profit from the income statement
revenue = income_statement.loc['Total Revenue']
net_income = income_statement.loc['Net Income']

# Plot revenue and profit trends
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(revenue.index, revenue.values, label='Total Revenue')
plt.plot(net_income.index, net_income.values, label='Net Income', linestyle='--')
plt.title('Starbucks Revenue and Profit Trends')
plt.xlabel('Date')
plt.ylabel('Amount (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming income_statement is already loaded and has the correct data

# Extract expense data
interest_expense = income_statement.loc[:, 'Interest Expense']
general_administrative_expense = income_statement.loc[:, 'General And Administrative Expense']
rent_expense = income_statement.loc[:, 'Rent Expense Supplemental']

# Plotting
plt.figure(figsize=(14, 10))

# Interest Expense
plt.subplot(3, 1, 1)
plt.plot(interest_expense.index, interest_expense.values, marker='o', linestyle='-', color='b')
plt.title('Interest Expenses Over Time')
plt.xlabel('Date')
plt.ylabel('Interest Expense (USD)')
plt.grid(True)

# General and Administrative Expenses
plt.subplot(3, 1, 2)
plt.plot(general_administrative_expense.index, general_administrative_expense.values, marker='o', linestyle='-', color='g')
plt.title('General and Administrative Expenses Over Time')
plt.xlabel('Date')
plt.ylabel('General and Administrative Expense (USD)')
plt.grid(True)

# Rent Expenses Supplemental
plt.subplot(3, 1, 3)
plt.plot(rent_expense.index, rent_expense.values, marker='o', linestyle='-', color='r')
plt.title('Rent Expenses Supplemental Over Time')
plt.xlabel('Date')
plt.ylabel('Rent Expense Supplemental (USD)')
plt.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

3. Key Financial Ratios:

```python
# Extract EPS and price data
current_price = starbucks_data['Close'].iloc[-1]  # Latest closing price
trailing_eps = starbucks.info.get('trailingEps', 0)  # Trailing EPS
```

```python
# Calculate P/E Ratio
pe_ratio = current_price / trailing_eps 
```

```python
print(f"Current Price: {current_price}")
print(f"Earnings Per Share (EPS): {trailing_eps}")
print(f"P/E Ratio: {pe_ratio}")
```
